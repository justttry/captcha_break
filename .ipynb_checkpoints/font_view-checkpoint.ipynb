```python
import random
import os
from itertools import product
from PIL import Image, ImageDraw, ImageFont
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
chars = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabdefghijlmnqrtuwxy"
font_paths = []
font_dir = 'fonts/english'
for dirpath, dirnames, filenames in os.walk(font_dir):
    for filename in filenames:
        filepath = dirpath + os.sep + filename
        font_paths.append({'eng':filepath, 'name':filename})
```

```python
for fontpath in font_paths:
    im = Image.new(mode='RGBA', size=(5+20*11, 5+20*5))
    fill = (111, 111, 254)
    size_cha = 20
    font = ImageFont.truetype(fontpath['eng'], size_cha)
    drawer = ImageDraw.Draw(im)
    for i, char in enumerate(chars):
        drawer.text(xy=(5+20*(i%11), 5+20*(i/11)), 
                    text=char, 
                    fill=fill, 
                    font=font)
    plt.imshow(im)
    plt.xlabel(fontpath['name'])
    plt.show()
```
